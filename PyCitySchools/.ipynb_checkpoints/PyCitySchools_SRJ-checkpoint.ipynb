```python
#import dependencies
import pandas as pd
```

```python
#sets up dataframes
school_data = "Resources/schools_complete.csv"
student_data = "Resources/students_complete.csv"
school_data_df = pd.read_csv(school_data, encoding="ISO-8859-1")
student_data_df = pd.read_csv(student_data, encoding="ISO-8859-1")
```

```python
school_data_df.head()
```

```python
student_data_df.head()
```

```python
#merges two dataframes
school_data_complete = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#District Summary
#number of schools
totSchool_l = list(school_data_complete['school_name'].unique())
totSchool_n=len(totSchool_l)
#total students
totStudents=len(school_data_complete['school_name'])
#total budget
totbudget= sum(school_data_df['budget'])
#average math and reading score
avgmath =school_data_complete["math_score"].mean()
avgreading = school_data_complete["reading_score"].mean()
# % passing math and reading 
mathPass_df = school_data_complete.loc[school_data_complete["math_score"]>=70]
mathPass_n=len(mathPass_df)
mathPer = (mathPass_n/totStudents)*100
readingPass_df = school_data_complete.loc[school_data_complete["reading_score"]>=70]
readingPass_n=len(readingPass_df)
readingPer = (readingPass_n/totStudents)*100
#overall passing
overallPass_df = mathPass_df.loc[mathPass_df["reading_score"]>=70]
overallPass_n = len(overallPass_df)
overallPer = (overallPass_n/totStudents)*100

district_df = pd.DataFrame([{"Number of Schools": totSchool_n,
                            "Total Students": totStudents, 
                            "Total Budget": totbudget, 
                            "Average Math Score ": avgmath, 
                            "Average Reading Score": avgreading, 
                            "% Passing Math": mathPer, 
                            "% Passing Reading": readingPer, 
                            "% Overall Passing": overallPer}])
district_df["Total Budget"] = district_df["Total Budget"].astype(float).map(
    "${:,.2f}".format)
district_df
```

```python
#School Summary 
    #creating the dataframe with the first four columns for each school
    
#determine school type
school_types = school_data_df.set_index(["school_name"])["type"]
types_df = school_types.to_frame()
types_df = types_df.sort_values(by = ['school_name'])
#calc total student count
per_school_counts = school_data_complete["school_name"].value_counts()
per_school_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]
per_school_capita = per_school_budget / per_school_counts

#calc avg test scores
per_school_math = school_data_complete.groupby(["school_name"]).mean()["math_score"]
per_school_reading = school_data_complete.groupby(["school_name"]).mean()["reading_score"]

#find out who is passing
school_passing_math = school_data_complete[(school_data_complete["math_score"]>=70)]
school_passing_reading = school_data_complete[(school_data_complete["reading_score"]>=70)]
school_passing_overall = school_passing_math[(school_passing_math["reading_score"]>=70)]

#how many per school
school_passing_math_n = school_passing_math["school_name"].value_counts()
school_passing_reading_n = school_passing_reading["school_name"].value_counts()
school_passing_overall_n = school_passing_overall["school_name"].value_counts()
percent_school_math = (school_passing_math_n/per_school_counts)*100
percent_school_reading = (school_passing_reading_n/per_school_counts)*100
percent_school_overall = (school_passing_overall_n/per_school_counts)*100

types_df ["Total Students"] = per_school_counts
types_df ["Total School Budget"] = per_school_budget
types_df ["Budget Per Student"] = per_school_capita
types_df ["Average Math Score"] = per_school_math
types_df ["Average Reading Score"] = per_school_reading
types_df ["% Passing Math"] = percent_school_math
types_df ["% Passing Reading"] = percent_school_reading
types_df ["% Passing Overall"] = percent_school_overall

types2 = types_df.copy()
types_df
```

```python
types3 = types_df.copy()
types3["Total School Budget"] = types3["Total School Budget"].astype(float).map(
    "${:,.2f}".format)
types3["Budget Per Student"] = types3["Budget Per Student"].astype(float).map(
    "${:,.2f}".format)
types3
```

```python
#math score by grade
school_grade = school_data_df.set_index(["school_name"])["type"]
by_grade_math = school_grade.to_frame()
by_grade_math = by_grade_math.sort_values(by = ['school_name'])
grade_9 = school_data_complete[(school_data_complete["grade"]=="9th")]
grade_10 = school_data_complete[(school_data_complete["grade"]=="10th")]
grade_11 = school_data_complete[(school_data_complete["grade"]=="11th")]
grade_12 = school_data_complete[(school_data_complete["grade"]=="12th")]
math_9 = grade_9.groupby(["school_name"]).mean()["math_score"]
math_10 = grade_10.groupby(["school_name"]).mean()["math_score"]
math_11 = grade_11.groupby(["school_name"]).mean()["math_score"]
math_12 = grade_12.groupby(["school_name"]).mean()["math_score"]
by_grade_math ["9th"] = math_9
by_grade_math ["10th"] = math_10
by_grade_math ["11th"] = math_11
by_grade_math ["12th"] = math_12
by_grade_math = by_grade_math.loc[:, ["9th", "10th", "11th",
                                    "12th"]]
by_grade_math
```

```python
#reading score by grade
school_grade = school_data_df.set_index(["school_name"])["type"]
by_grade_reading = school_grade.to_frame()
by_grade_reading = by_grade_reading.sort_values(by = ['school_name'])
grade_9 = school_data_complete[(school_data_complete["grade"]=="9th")]
grade_10 = school_data_complete[(school_data_complete["grade"]=="10th")]
grade_11 = school_data_complete[(school_data_complete["grade"]=="11th")]
grade_12 = school_data_complete[(school_data_complete["grade"]=="12th")]
reading_9 = grade_9.groupby(["school_name"]).mean()["reading_score"]
reading_10 = grade_10.groupby(["school_name"]).mean()["reading_score"]
reading_11 = grade_11.groupby(["school_name"]).mean()["reading_score"]
reading_12 = grade_12.groupby(["school_name"]).mean()["reading_score"]
by_grade_reading ["9th"] = reading_9
by_grade_reading ["10th"] = reading_10
by_grade_reading ["11th"] = reading_11
by_grade_reading ["12th"] = reading_12
by_grade_reading
```

```python
#Sorted by Size of School

# Create bins in which to place values based upon School Summary
bins = [0, 1000, 2000, 5000]

# Create labels for these bins
group_labels = ["Small(>1000)", "Medium(1000-2000)", "Large(2000-5000)"]

# Slice the data and place it into bins
pd.cut(types2["Total Students"], bins, labels=group_labels)
# Place the data series into a new column inside of the DataFrame
types2["Size Group"] = pd.cut(types2["Total Students"], bins, labels=group_labels)
types2.head()
# Create a GroupBy object based upon "Size Group"
size_group = types2.groupby("Size Group")
# Find how many rows fall into each bin
size_n = (size_group["Total Students"].count())
# Get the average of each column within the GroupBy object
size_df = size_group[["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Passing Overall"]].mean()
size_df
```

```python
#Sorted by Budget per Student
# Create bins in which to place values based upon TED Talk views
budget_bins = [0, 584, 629, 644, 675]

# Create labels for these bins
budget_labels = ["<$584", "$585-629", "$630-644", "$645-675"]
# Place the data series into a new column inside of the DataFrame
types2["Budget Group"] = pd.cut(types2["Budget Per Student"], budget_bins, labels=budget_labels)
# Create a GroupBy object based upon "Size Group"
budget_group = types2.groupby("Budget Group")
budget_n = (budget_group["Budget Group"].count())
budget_n
budget_df = budget_group[["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Passing Overall"]].mean()
budget_df
```

```python
#Scores by School Type
math_type = types_df.groupby(["type"])["Average Math Score"].mean()
reading_type = types_df.groupby(["type"])["Average Reading Score"].mean()
mathpass_type = types_df.groupby(["type"])["% Passing Math"].mean()
readingpass_type = types_df.groupby(["type"])["% Passing Reading"].mean()
overallpass_type = types_df.groupby(["type"])["% Passing Overall"].mean()


school_type_df= pd.DataFrame({"Average Math Score": math_type,
                             "Average Reading Score": reading_type,
                             "% Passing Math": mathpass_type,
                             "% Passing Reading": readingpass_type,
                             "% Passing Overall": overallpass_type})
school_type_df
```
